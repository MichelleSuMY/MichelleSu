```python
# Question 3
# Below is the improved and simplified Monty Hall Simulattion code generated with the help of ChatGPT

import numpy as np

def monty_hall_simulation(reps=100000, initial_choice=1):
    wins = 0
    doors = [1, 2, 3]

    for _ in range(reps):
        # Randomly select the winning door and the player's initial choice
        winning_door = np.random.choice(doors) 
        # The new variable names are simpler and more comprehensive here. Instead of secret_winning_door, it is winning_door
        
        # Determine the door that is revealed (a goat)
        remaining_doors = [door for door in doors if door != winning_door and door != initial_choice]
        revealed_door = np.random.choice(remaining_doors)

        # Determine the new choice after swapping
        new_choice = next(door for door in doors if door != initial_choice and door != revealed_door)

    # The new code disposes of the try-except statement in the original for brevity
        
        # Check if the new choice is the winning door
        if new_choice == winning_door:
            wins += 1
    
            
    return wins / reps

# Execute the simulation
win_probability = monty_hall_simulation()
print(f"Win probability when swapping: {win_probability:.4f}")

# Overall, the new code is shorter and more comprehensive with less complex steps compared to the original code
# There are also more boolean operators used in the new code
```
