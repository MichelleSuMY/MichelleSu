# Proposal 1
### Research Question

Does tendency toward altruism correlate with increased loneliness in students?
<br>
### Variables
#### Independent Variable
Altruism will be measured by the "PSYCH_big_five_inventory_helpful" variable, representing the degree of agreement with the statement, "I see myself as someone who is helpful and not selfish with others". This measure represents altruism as it reflects students' tendency to consider others' needs.
#### Dependent Variable
Loneliness will be measured by the "LONELY_dejong_emotional_social_loneliness_scale_emptiness" variable, representing the degree of agreement with the statement, "I experience a general sense of emptiness", which serves as a loneliness indicator.
#### Other Variables
Student status will be filtered using the "DEMO_student" variable to ensure only student responses are included.
#### Visualizations
Each variable will be visualized using histograms to show the distribution of responses, and a scatter plot will visualize the relationship between altruism and loneliness, helping to identify any trends or patterns.
<br>

### Analysis

Bootstrapping will be used to simulate population parameters due to a small sample size. Then, a correlation test will be conducted to analyze the relationship between altruism and loneliness. Following this, a hypothesis test will be performed to examine whether altruistic tendencies affect loneliness levels. Simple linear regression will also be applied to assess the trend between altruism and loneliness, with assumptions such as linearity and independence evaluated.
<br>

### Hypothesis
Students who score higher on altruism measures are less likely to feel lonely. If we reject the null hypothesis, this will indicate a significant relationship between altruism and loneliness. The simple linear regression results will illustrate the trend and strength of this relationship, providing insights into the psychological impacts of altruistic behavior on student well-being.

# Proposal 2
### Research Question

Are students with learning disabilities more likely to seek help from colleages than students without learning disabilities?

<br>

### Variables
#### Independent Variable
The presence of learning disability in students will be measured by the "DEMO_disability_learning" variable, which is binary (students with or without learning disabilities).
#### Dependent Variable
The tendency to seek help from colleagues will be measured by the "SOCIAL_ask_help_colleague" variable, which uses an ordinal scale to indicate the likelihood of asking colleagues for help when needed.
#### Other Variables
Student status will be filtered using the "DEMO_student" variable to ensure only student responses are included.
#### Visualizations
The distribution of responses to the dependent variable will be visualized with a histogram to understand the overall trend in help-seeking behaviors. A box plot or violin plot will compare help-seeking tendencies between students with and without learning disabilities, providing insights into central tendency and variability across groups.

<br>

### Analysis
Bootstrapping will be used to estimate population parameters due to limited sample size. An independent samples t-test or Mann-Whitney U test will be performed to compare help-seeking behaviors between students with and without learning disabilities, as these are appropriate for binary and ordinal data comparisons. Then, logistic regression may be applied to model the likelihood of seeking help based on learning disability status, offering insights into whether learning disabilities increase the odds of asking for help. Assumptions of normality (for the t-test) and independence will be evaluated as appropriate.
<br>

### Hypothesis
Students with learning disabilities are statistically more likely to seek help from colleagues than students without learning disabilities. If we reject the null hypothesis, this suggests a significant relationship between learning disabilities and help-seeking behavior, potentially indicating a unique need for social support among students with learning disabilities.
